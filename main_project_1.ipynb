# imports

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random
```

# reading files


```python
csv1 = 'https://raw.githubusercontent.com/Taldafna/High-School-Student/main/student_math_clean.csv'
csv2 = 'https://raw.githubusercontent.com/Taldafna/High-School-Student/main/student_portuguese_clean.csv'
math_df = pd.read_csv(csv1)
portuguese_df = pd.read_csv(csv2)
```

# combine the files

```python
portuguese_df['course'] = 'portuguese'
math_df['course'] = 'math'
```

```python
Student_Performance_df = pd.concat([math_df, portuguese_df]).reset_index(drop=True)
Student_Performance_df.drop('student_id', axis=1, inplace=True)
```

# adding, replacing data and correlation

```python
# הוספת עמודה שבה קטגוריה של הציון הסופי
not_exellent_df = Student_Performance_df.copy()
Student_Performance_df['grade_category'] = pd.cut(
    Student_Performance_df['final_grade'],
    bins=[0, 10, 15, 20],
    labels=['weak', 'average', 'exellent'], # אחד הוא הגבוה
    include_lowest=True
)
```

> **binary columns**

```python
binary_columns = ['school_support', 'family_support', 'extra_paid_classes', 'activities', 'nursery_school', 'higher_ed', 'internet_access', 'romantic_relationship']
for column in binary_columns:
    print(f"Unique values in {column}: {Student_Performance_df[column].unique()}")
```

```python
binary_maping = {'yes': 1, 'no': 0}
for column in binary_columns:
    Student_Performance_df[column] = Student_Performance_df[column].map(binary_maping)
```

<p style="direction: rtl; text-align: right;">
העברנו את העמודות שהערך שלהן הוא  yes, no לערכי 0,1 .
</p>

> srting columns

```python
int_only_df = Student_Performance_df.copy()
categorical_columns = ['family_size', 'parent_status', 'mother_education', 'father_education', 'mother_job', 'father_job', 'school_choice_reason', 'travel_time', 'study_time', 'address_type', 'guardian', 'course']

for column in categorical_columns:
    int_only_df[column] = int_only_df[column].astype('category').cat.codes
```

<p style="direction: rtl; text-align: right;">
 הפכנו את כל העמודות שיש בהן נתון מילולי לקטגורי למקרה שנרצה לעשות קוראלציה בהמשך
</p>

> # full corelation map

```python
plt.figure(figsize=(20,8))
performance_corr_df = int_only_df.corr(numeric_only = True, method = 'pearson').round(2)
sns.heatmap(performance_corr_df,cmap="Pastel2",linecolor ="black",linewidth = "0.5",xticklabels=performance_corr_df.columns, yticklabels=performance_corr_df.columns , annot=True)
plt.xticks(rotation=25, ha='right')
plt.yticks(rotation=0)
plt.show()
```
